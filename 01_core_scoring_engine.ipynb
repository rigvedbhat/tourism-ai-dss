# System Philosophy — Core Variables in the Tourism AI DSS

This Decision Support System (DSS) models tourism flow using three core variables: **Congestion**, **Attractiveness**, and **Feasibility**, which are synthesized into a unified **Match Score**.

All scores in this prototype are normalized to a **0–100 scale** for interpretability and operational clarity.

The AI ranks alternatives. The human decides.

---

## 1. Congestion (Overtourism Index)

Congestion is derived from tourism carrying capacity theory and the Tourism Area Life Cycle (TALC) framework.

It represents the condition where tourist inflow approaches or exceeds the physical, infrastructural, or social limits of a destination.

### In this system, congestion reflects:
- Load ratio (tourist_count / capacity)
- Travel delay and crowd-induced friction
- Risk of degraded visitor experience

Congestion is modeled as a **negative externality** affecting both:
- Local resident welfare
- Tourist psychological satisfaction

It is scaled from **0 to 100**, where:
- 0 = No congestion
- 100 = Severe overtourism pressure

The DSS aims to minimize excessive imbalance rather than eliminate congestion entirely.

---

## 2. Attractiveness (Pull Potential)

Attractiveness is based on destination attractiveness literature and represents a location’s perceived ability to satisfy tourist motivations.

It is a multidimensional construct including:
- Cultural or historical value
- Experiential richness
- Infrastructure quality
- Overall desirability

Attractiveness acts as the psychological anchor.

An alternative (Orbit) must maintain sufficient attractiveness to be considered comparable to a hotspot (Anchor).

It is scaled from **0 to 100**, where:
- 0 = No experiential value
- 100 = Extremely high pull potential

---

## 3. Feasibility (Operational Penalty Layer)

Feasibility is grounded in operations research and Decision Support System theory.

It evaluates whether a recommendation is practically executable.

Rather than acting as a hard constraint, feasibility is implemented as a **penalty deduction** from the final score.

### Examples:
- No coach parking for large groups
- Inadequate group dining facilities
- Excessive walking distance
- Infrastructure mismatch for tour size

Penalty values are also scaled within a **0–100 framework**.

This ensures the system does not unrealistically promote locations that are operationally unsuitable.

---

## 4. Match Score (Unified Ranking Metric)

The Match Score is the primary output of the system.

It aggregates competing objectives into a single interpretable metric.

### Formula (Phase 1 Prototype):

Match Score = Attractiveness − Congestion − Penalty

All components are normalized to the 0–100 scale.

Higher Match Score → More suitable recommendation.

This score does not enforce a decision.
It provides structured ranking support.

---

## 5. Human-in-the-Loop (HITL) Governance

This system follows a Human-in-the-Loop (HITL) architecture.

The AI:
- Calculates congestion and attractiveness
- Applies operational penalties
- Generates ranked alternatives
- Produces warning flags

The human operator:
- Interprets context
- Evaluates business realities
- Makes the final decision

This design mitigates automation bias and ensures the system augments professional judgment rather than replacing it.

## Congestion Index Design

````python
df["load_ratio"] = df["tourist_count"] / df["capacity"]

df[["region", "tourist_count", "capacity", "load_ratio"]]
````

````python
def absolute_congestion_score(ratio):
    if ratio < 0.3:
        return ratio * (20 / 0.3)  # Scale 0–20
    elif ratio < 0.7:
        return 20 + (ratio - 0.3) * (30 / 0.4)  # Scale 20–50
    elif ratio < 0.9:
        return 50 + (ratio - 0.7) * (25 / 0.2)  # Scale 50–75
    elif ratio < 1.0:
        return 75 + (ratio - 0.9) * (15 / 0.1)  # Scale 75–90
    elif ratio < 1.2:
        return 90 + (ratio - 1.0) * (10 / 0.2)  # Scale 90–100
    else:
        return 100

df["load_score"] = df["load_ratio"].apply(absolute_congestion_score)

df[["region", "load_ratio", "load_score"]]
````

## Phase 1 Full Scoring Engine

The following cells implement the complete scoring pipeline across all four components:

1. **Congestion Index** — Upgraded piecewise load score (no ceiling), delay score, and Dominance Rule fusion
2. **Attractiveness Score** — Weighted composite of cultural, experiential, and infrastructure dimensions
3. **Feasibility Penalty** — Additive operational deductions for group tour suitability
4. **Match Score** — Final unified ranking metric

> *Theoretical grounding: The congestion quadratic term mirrors Asahi & Habu (2025) — load ratio exceeding carrying capacity (CC) is modelled as a nonlinear negative externality, consistent with the TALC demand framework used in their dynamic panel GMM estimates for Kyoto City.*

### Step 1 — Congestion Index (Upgraded)

**Changes from prototype:**
- The `>= 1.2` branch now uses `100 + ((ratio - 1.2) * 20)` — no ceiling, to expose the full severity of overtourism pressure (e.g. Kyoto at load_ratio 1.5 becomes 106)
- `delay_score` is added as a secondary congestion signal
- **Dominance Rule**: when `load_score >= 100`, delay is treated as additive amplification (×0.20) rather than a blended signal

````python
import numpy as np

# ── 1a. Upgraded load_score (no ceiling above 1.2) ─────────────────────────
def piecewise_load_score(ratio):
    """Piecewise congestion score. No ceiling for ratio >= 1.2."""
    if ratio < 0.3:
        return ratio * (20 / 0.3)
    elif ratio < 0.7:
        return 20 + (ratio - 0.3) * (30 / 0.4)
    elif ratio < 0.9:
        return 50 + (ratio - 0.7) * (25 / 0.2)
    elif ratio < 1.0:
        return 75 + (ratio - 0.9) * (15 / 0.1)
    elif ratio < 1.2:
        return 90 + (ratio - 1.0) * (10 / 0.2)
    else:
        return 100 + ((ratio - 1.2) * 20)   # No ceiling — exposes full severity

df["load_score"] = df["load_ratio"].apply(piecewise_load_score)

# ── 1b. Delay score (clipped at 100) ───────────────────────────────────────
df["delay_score"] = (df["avg_delay_minutes"] / 45 * 100).clip(upper=100)

# ── 1c. Congestion Index — Dominance Rule ──────────────────────────────────
def congestion_index(row):
    """Dominance Rule: when load_score >= 100, delay amplifies rather than blends."""
    if row["load_score"] >= 100:
        return row["load_score"] + (row["delay_score"] * 0.20)
    else:
        return (row["load_score"] * 0.70) + (row["delay_score"] * 0.30)

df["congestion_index"] = df.apply(congestion_index, axis=1)

# ── 1d. Congestion flag ────────────────────────────────────────────────────
def congestion_flag(index):
    if index < 20:
        return "Severely undertouristed"
    elif index < 40:
        return "Undertouristed"
    elif index < 75:
        return "Balanced"
    elif index < 100:
        return "Approaching overtourism"
    else:
        return "Severely overtouristed"

df["congestion_flag"] = df["congestion_index"].apply(congestion_flag)

df[["region", "load_ratio", "load_score", "delay_score", "congestion_index", "congestion_flag"]]
````

### Step 2 — Attractiveness Score

Weighted composite of three destination quality dimensions.

Cultural and experiential dimensions carry equal weight (0.40 each) as primary visitor motivators, consistent with heritage tourism literature. Infrastructure acts as an enabling factor (0.20).

> *Note: `promotion_score` is retained in the dataset for future phases (e.g. marketing uplift modelling) but is excluded from Phase 1 attractiveness to avoid conflating perceived quality with promotional visibility.*

````python
# ── 2. Attractiveness Score ────────────────────────────────────────────────
df["attractiveness_score"] = (
    (df["cultural_score"]       * 0.40) +
    (df["experiential_score"]   * 0.40) +
    (df["infrastructure_score"] * 0.20)
)

df[["region", "cultural_score", "experiential_score", "infrastructure_score", "attractiveness_score"]]
````

### Step 3 — Feasibility Penalty Score

Additive operational deductions simulating a **large tour group** context.

| Condition | Penalty |
|---|---|
| No coach parking | −20 |
| No group dining nearby | −15 |
| Requires long walk | −10 |

Penalties are additive. Maximum possible penalty = 45.

````python
# ── 3. Feasibility Penalty Score ───────────────────────────────────────────
df["penalty_score"] = (
    (~df["has_coach_parking"]).astype(int)  * 20 +
    (~df["group_dining_nearby"]).astype(int) * 15 +
    (df["requires_long_walk"]).astype(int)   * 10
)

df[["region", "has_coach_parking", "group_dining_nearby", "requires_long_walk", "penalty_score"]]
````

### Step 4 — Final Match Score & Ranked Output

**Formula:**
```
match_score = attractiveness_score - (congestion_index × 0.5) - penalty_score
```

The `congestion_index` is halved (×0.5) to balance it against attractiveness — congestion is a strong negative signal, but should not fully override destination quality for lightly penalised Orbit destinations.

A higher Match Score → more suitable recommendation for tour diversion away from overtouristed Anchors.

````python
import pandas as pd
import numpy as np

# ── 4. Final Match Score ───────────────────────────────────────────────────
df["match_score"] = (
    df["attractiveness_score"]
    - (df["congestion_index"] * 0.5)
    - df["penalty_score"]
).round(2)

# ── Ranked summary output ──────────────────────────────────────────────────
DISPLAY_COLS = [
    "region",
    "is_anchor",
    "congestion_flag",
    "attractiveness_score",
    "penalty_score",
    "match_score",
]

results = (
    df[DISPLAY_COLS]
    .sort_values("match_score", ascending=False)
    .reset_index(drop=True)
)

# Style for legibility
results.style \
    .format({"attractiveness_score": "{:.1f}", "match_score": "{:.2f}"}) \
    .background_gradient(subset=["match_score"], cmap="RdYlGn") \
    .map(
        lambda v: "color: crimson; font-weight: bold"
            if v == "Severely overtouristed"
        else ("color: darkorange" if v == "Approaching overtourism" else ""),
        subset=["congestion_flag"]
    ) \
    .set_caption(
        "Phase 1 Match Score Rankings — Tourism AI DSS "
        "(sorted by match_score DESC)"
    )
````
