# System Philosophy — Core Variables in the Tourism AI DSS

This Decision Support System (DSS) models tourism flow using three core variables: **Congestion**, **Attractiveness**, and **Feasibility**, which are synthesized into a unified **Match Score**.

All scores in this prototype are normalized to a **0–100 scale** for interpretability and operational clarity.

The AI ranks alternatives. The human decides.

---

## 1. Congestion (Overtourism Index)

Congestion is derived from tourism carrying capacity theory and the Tourism Area Life Cycle (TALC) framework.

It represents the condition where tourist inflow approaches or exceeds the physical, infrastructural, or social limits of a destination.

### In this system, congestion reflects:
- Load ratio (tourist_count / capacity)
- Travel delay and crowd-induced friction
- Risk of degraded visitor experience

Congestion is modeled as a **negative externality** affecting both:
- Local resident welfare
- Tourist psychological satisfaction

It is scaled from **0 to 100**, where:
- 0 = No congestion
- 100 = Severe overtourism pressure

The DSS aims to minimize excessive imbalance rather than eliminate congestion entirely.

---

## 2. Attractiveness (Pull Potential)

Attractiveness is based on destination attractiveness literature and represents a location’s perceived ability to satisfy tourist motivations.

It is a multidimensional construct including:
- Cultural or historical value
- Experiential richness
- Infrastructure quality
- Overall desirability

Attractiveness acts as the psychological anchor.

An alternative (Orbit) must maintain sufficient attractiveness to be considered comparable to a hotspot (Anchor).

It is scaled from **0 to 100**, where:
- 0 = No experiential value
- 100 = Extremely high pull potential

---

## 3. Feasibility (Operational Penalty Layer)

Feasibility is grounded in operations research and Decision Support System theory.

It evaluates whether a recommendation is practically executable.

Rather than acting as a hard constraint, feasibility is implemented as a **penalty deduction** from the final score.

### Examples:
- No coach parking for large groups
- Inadequate group dining facilities
- Excessive walking distance
- Infrastructure mismatch for tour size

Penalty values are also scaled within a **0–100 framework**.

This ensures the system does not unrealistically promote locations that are operationally unsuitable.

---

## 4. Match Score (Unified Ranking Metric)

The Match Score is the primary output of the system.

It aggregates competing objectives into a single interpretable metric.

### Formula (Phase 1 Prototype):

Match Score = Attractiveness − Congestion − Penalty

All components are normalized to the 0–100 scale.

Higher Match Score → More suitable recommendation.

This score does not enforce a decision.
It provides structured ranking support.

---

## 5. Human-in-the-Loop (HITL) Governance

This system follows a Human-in-the-Loop (HITL) architecture.

The AI:
- Calculates congestion and attractiveness
- Applies operational penalties
- Generates ranked alternatives
- Produces warning flags

The human operator:
- Interprets context
- Evaluates business realities
- Makes the final decision

This design mitigates automation bias and ensures the system augments professional judgment rather than replacing it.

## Congestion Index Design

```python
df["load_ratio"] = df["tourist_count"] / df["capacity"]

df[["region", "tourist_count", "capacity", "load_ratio"]]
```

```python
def absolute_congestion_score(ratio):
    if ratio < 0.3:
        return ratio * (20 / 0.3)  # Scale 0–20
    elif ratio < 0.7:
        return 20 + (ratio - 0.3) * (30 / 0.4)  # Scale 20–50
    elif ratio < 0.9:
        return 50 + (ratio - 0.7) * (25 / 0.2)  # Scale 50–75
    elif ratio < 1.0:
        return 75 + (ratio - 0.9) * (15 / 0.1)  # Scale 75–90
    elif ratio < 1.2:
        return 90 + (ratio - 1.0) * (10 / 0.2)  # Scale 90–100
    else:
        return 100

df["load_score"] = df["load_ratio"].apply(absolute_congestion_score)

df[["region", "load_ratio", "load_score", "congestion_state"]]
```
